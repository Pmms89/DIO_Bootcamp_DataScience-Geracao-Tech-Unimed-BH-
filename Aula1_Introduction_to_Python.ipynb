### Working with strings, but the same idea can be used with other kinds of data (e.g. float, int, and so on):

```python
# How to create a string variable:
phrase = "Python to data analysis"
```

```python
# Getting the first element of the variable phrase
phrase[0]
```

```python
# Getting the first nine elements of the variable phrase
phrase[0:10]
```

```python
# Getting the tenth until the last element of the variable phrase
phrase[10:]
```

```python
# Getting the first until the seventeenth element of the variable phrase
phrase[:17]
```

```python
# Counting how many times the string "data" appear in the variable phrase
phrase.count("data")
```

```python
# Putting all elements of the varaible phrase in lower case
phrase.lower()
```

```python
# Putting all elements of the varaible phrase in upper case and adding it the variable
phrase = phrase.upper()
print(phrase)
```

```python
# Replacing elements inside the variable phrase
phrase.replace("PYTHON","Phrases")
```

```python
# Showing the variable phrase
phrase
```
