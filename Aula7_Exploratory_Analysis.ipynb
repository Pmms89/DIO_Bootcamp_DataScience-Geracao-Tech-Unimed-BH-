```python
# Creating our dataframe
df = pd.read_excel("/Users/patriciasilva/Downloads/Dio_Unimed_Bootcamp/Cusro_Python_Pandas_Digital_Innovation-master/datasets/AdventureWorks.xlsx")
```

```python
# Looking the first five rows
df.head()
```

```python
# Quantity of rows and columns
df.shape
```

```python
# Verifying data types
df.dtypes
```

```python
# What is the total revenue
df["Valor Venda"].sum()
```

```python
# Creating the column custo
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) 
```

```python
df.head(1)
```

```python
# What is the total cost?
round(df["custo"].sum(), 2)
```

```python
# With the revenue and total cost, we can find the profit
# Creating the column profit (revenue - cost)
df["lucro"]  = df["Valor Venda"] - df["custo"] 
```

```python
df.head(1)
```

```python
# Total profit - rounding to two decimal places
round(df["lucro"].sum(),2)
```

```python
# Creating a column with total of days to send the product
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

### To know the mean time to send for each brand

For this we neew to transform the column "Tempo_envio" in numerical

```python
# Extracting only the days
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
# Verifying the date tyoe of the column Tempo_envio
df["Tempo_envio"].dtype
```

```python
# Mean of time to send by brand
df.groupby("Marca")["Tempo_envio"].mean()
```

 ### **Treating Missing Values**

```python
# Verifying missing values
df.isnull().sum()
```

### If we want to see the profit per year and per brand?

```python
# Grouping by year and brand
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
 pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Reseting the index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
# What is the total product sold?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Graph of Total Products sold
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Products sold")
plt.xlabel("Total")
plt.ylabel("Product");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Profit per year")
plt.xlabel("Year")
plt.ylabel("Profit");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
# Selecting only 2009 sales 
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Profit per month")
plt.xlabel("Month")
plt.ylabel("Profit");
```

```python
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Profit per Brand")
plt.xlabel("Brand")
plt.ylabel("Profit")
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Profit per category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation='horizontal');
```
