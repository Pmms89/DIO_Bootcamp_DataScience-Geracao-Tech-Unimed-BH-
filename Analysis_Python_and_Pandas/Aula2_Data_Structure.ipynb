# Data Structure

### **Lists**

```python
#Creating a list and calling it animals
animals = [1,2,3]
animals
```

```python
animals = ["dog", "cat", 12345, 6.5]
animals
```

```python
# Printing the first element of the list
animals[0]
```

```python
# Printing the fourth element of the list
animals[3]
```

```python
# Replacing the first element of the list
animals[0] = "parot"
```

```python
animals
```

```python
# Removing the element "cat" of the list
animals.remove("cat")
```

```python
animals
```

```python
# Seeing the length of the list - use the command len
len(animals)
```

```python
# Seeing if the element "cat" is in the list animals
"cat" in animals
```

```python
# Adding new elements in the list - in this case another list
animals.append(["lion", "dog"])
print(animals)
```

```python
# Adding new elements in the list - in this case separate elements
animals.extend(["snake", 6])
print(animals)
```

```python
# Seeing how many an element appear in a list - in this case lion isn't an individual element, because this
# element is inside another list
animals.count("lion")
```

```python
# Creating another list
newlist = [850, 30, 150, 80, 10]
```

```python
# Seeing the maximum value in the list
max(newlist)
```

```python
# Seeing the minimum value in the list
min(newlist)
```

```python
# Sorting the list in ascending order
newlist.sort()
print(newlist)
```

```python
# Sorting the list in descending order
newlist.sort(reverse=True)
print(newlist)
```

### **Tuples**

Tuples different from lists are defined using parenthesis () and once discriminated it can't be altered

```python
# Tuples defined using parenthesis
tp = ("Banana", "Apple", 10, 50)
```

```python
# Seeing the first element of the tuple
tp[0]
```

```python
# Tuples are immuable, you can not alter its elements - if you try it will return an error
# (TypeError: 'tuple' object does not support item assignment)
tp[0] = "Orange"
```

```python
# Seeing how many an element appear in a tuple
tp.count("Apple")
```

```python
# Seeing the first and second element
tp[0:2]
```

### **Dictionaries**

Dictionaries are an unordered set of key:value pairs, where the keys are unique in a given dictionary instance. Dictionaries are delimited by square brackets: {}, and contain a comma-separated list of key:value pairs.

```python
# To create a dictionary we use square brackets {}
# Dictionaries working with the concept key:value pairs
dc = {"Apple":20, "Banana":10, "Orange":15, "Grape":5} 
```

```python
dc
```

```python
# Accessing a dictionary value using its key
dc["Apple"]
```

```python
# Updating "Apple" value
dc["Apple"] = 25
dc
```

```python
# Returning all dictionary keys
dc.keys()
```

```python
# Returning all dictionary values
dc.values()
```

```python
# Verifying if one key already exist in a dictionary, in case of nonexistence insert it
dc.setdefault("Lemon", 22)
```
