### **Treating missing values**

Different forms to treat missing values

```python
# Consulting rows with missing values
df.isnull().sum()
```

```python
# Replacing the null values for the mean
df["Sales"].fillna(df["Sales"].mean(), inplace=True)
```

```python
df["Sales"].mean()
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
# Replacing null values for zero
df["Sales"].fillna(0, inplace=True)
```

```python
# Removing rows with null values
df.dropna(inplace=True)
```

```python
# Removing the rows with null values based only in one columns
df.dropna(subset=["Sales"], inplace=True)
```

```python
# Removing rows with null values in all columns
df.dropna(how="all", inplace=True)
```

### **Creating new columns**

```python
# Creating the column revenue
df["Revenue"] = df["Sales"].mul(df["Qty"])
```

```python
df["Revenue/Sales"] = df["Revenue"] / df["Sales"] 
```

```python
#nsamllest - Returning the nsmallest Revenues, in this example we put the 5 smallest
df.nsmallest(5, "Revenue")
```

### **Working with dates**

```python
# Transforming the date column in int type
df["Date"] = df["Date"].astype("int64")
```

```python
# Looking the data types in each column
df.dtypes
```

```python
# Transforming date column in date
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.dtypes
```

```python
# Grouping by date
df.groupby(df["Date"].dt.year)["Revenue"].sum()
```

```python
# Creating a new column with the year
df["Sales_year"] = df["Date"].dt.year
```

```python
df.sample(5)
```

```python
# Extracting the month and the day
df["Sales_month"], df["Sales_day"] = (df["Date"].dt.month, df["Date"].dt.day)
```

```python
df.sample(5)
```

```python
# Returning the most ancient date
df["Date"].min()
```

```python
# Calculating the difference of the days
df["days_difference"] = df["Date"] - df["Date"].min()
```

```python
df.sample(5)
```

```python
# Creating the quarter column
df["Sales_quarter"] = df["Date"].dt.quarter
```

```python
df.sample(5)
```

```python
# Filtering 2019 sales in march month
sales_march_19 = df.loc[(df["Date"].dt.year == 2019) & (df["Date"].dt.month == 3)]
```
